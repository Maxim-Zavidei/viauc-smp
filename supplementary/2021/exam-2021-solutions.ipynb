```python
# Null Hypothesis (H0): μ = 1
# Alternative Hypothesis (H1): μ ≠ 1
```

```python
# Sample data
sample = np.array([-1.7, 2.2, 3.2, -0.8, 2.0, 0.7, -2.2, 0.5, 3.0])

# Null hypothesis value
mu0 = 1  # Null hypothesis: the mean is 1

# Calculate sample statistics
n = len(sample)
x_bar = np.mean(sample)
s = np.std(sample, ddof=1)

# Calculate the test statistic
t_stat = (x_bar - mu0) / (s / np.sqrt(n))

print("Test statistic (t):", t_stat)
```

```python
# Significance level
alpha = 0.1

# Degrees of freedom
df = len(sample) - 1

# Calculate the critical value (two-tailed test)
critical_value = t.ppf(1 - alpha/2, df)

print("Critical value:", critical_value)
```

```python
# Exercise 10
Image('https://cdn.mathpix.com/snip/images/tM-fNVaDYMnMahbVzpdb5_-3mBFQpIFJ-2rI8ZbDmfw.original.fullsize.png', width = 700)
```

```python
x, y = smp.symbols('x y', real = True)

JOINT_PDF = R(1,2) * x**2 + R(2,3) * y

X_RANGE = (x, -1, 1)
Y_RANGE = (y, 0, 1)

print("Joint Probability Density Function:")
display(JOINT_PDF)

X_MPDF = smp.integrate(JOINT_PDF, Y_RANGE, conds = "none")
Y_MPDF = smp.integrate(JOINT_PDF, X_RANGE, conds = "none")

print("X Marginal PDF - PDF(X):")
display(X_MPDF)

print("Y Marginal PDF - PDF(Y):")
display(Y_MPDF)

# -------------------- Conditional PDFs --------------------

X_CONDITIONAL_PDF_GIVEN_Y = smp.simplify(JOINT_PDF / Y_MPDF)
print("X Conditional PDF Given Y - PDF(X | Y):")
display(X_CONDITIONAL_PDF_GIVEN_Y)

Y_CONDITIONAL_PDF_GIVEN_X = smp.simplify(JOINT_PDF / X_MPDF)
print("Y Conditional PDF Given X - PDF(Y | X):")
display(Y_CONDITIONAL_PDF_GIVEN_X)

print()

# Example: E[X | Y = 0]
E_X_GIVEN_Y = smp.integrate(x * smp.simplify(X_CONDITIONAL_PDF_GIVEN_Y.subs(y, 0)), X_RANGE)
print(f"{'E[X | Y = 0]:': <25}  {E_X_GIVEN_Y} , {round(float(E_X_GIVEN_Y), 2)}")

# Example: E[Y | X = 0]
E_Y_GIVEN_X = smp.integrate(y * smp.simplify(Y_CONDITIONAL_PDF_GIVEN_X.subs(x, 0)), Y_RANGE)
print(f"{'E[Y | X = 0]:': <25}  {E_Y_GIVEN_X} , {round(float(E_Y_GIVEN_X), 2)}")

print()
```

```python
x, y, c = symbols('x y c')

# Define the equations
eq1 = Eq(x - y - 1, 0)
eq2 = Eq(x - y + 1, 0)
eq3 = Eq(3 - y - x, 0)
eq4 = Eq(7 - y - x, 0)

# Solve the system of equations to find the points of intersection
intersection_points = solve((eq1, eq2, eq3, eq4), (x, y))

# Extract the x-coordinates of the intersection points
x_values = [point[x] for point in intersection_points]
y_values = [point[y] for point in intersection_points]

# Define the limits of integration
lower_x_limit = min(x_values)
upper_x_limit = max(x_values)

# Define the limits of integration
lower_y_limit = min(y_values)
upper_y_limit = max(y_values)

result = smp.solve(smp.integrate(c*x*y, (x, lower_x_limit, upper_x_limit), (y, lower_y_limit, upper_y_limit)) - 1, c)[0]
print("Exact result:", result)
result = nsimplify(result, tolerance=1e-10)
print(result)
```
